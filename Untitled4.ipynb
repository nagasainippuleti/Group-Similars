```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.feature_extraction.text import TfidfVectorizer
from gspread_dataframe import set_with_dataframe
import pandas as pd
import numpy as np
import datetime
import pygsheets
import nltk
import numpy as np
import string
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('omw-1.4')
nltk.download('wordnet')
import spacy
nlp = spacy.load('en_core_web_sm')



def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub('\d+', '', text)
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [t for t in tokens if not t in stop_words]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    text = ' '.join(tokens)
    return text


def extract_entities(text):
    doc = nlp(text)
    entities = [ent.text for ent in doc.ents if ent.label_ in ['PERSON', 'ORG', 'GPE', 'PRODUCT']]    # 'EVENT', 'WORK_OF_ART', 'LAW', 'LANGUAGE', 'DATE', 'QUANTITY', 'ORDINAL', 'CARDINAL'
    return ','.join(entities)

gc = pygsheets.authorize(service_file='python_sheets.json')
sheet = gc.open_by_key('1awXAxzgVQoU1hBCAbZ56mCaX4wZHtofp_N0dCMsR64s')

worksheet = sheet.worksheet_by_title('New Data')

news_titles = worksheet.get_col(3, include_tailing_empty=False)[1:]
titles = [preprocess(title) for title in news_titles]
entities = [extract_entities(title) for title in titles]

# Text similarity
vectorizer = TfidfVectorizer(stop_words='english')
X_text = vectorizer.fit_transform(titles)

# Entity similarity
vectorizer = TfidfVectorizer(stop_words='english')
X_entities = vectorizer.fit_transform(entities)

# Hybrid approach: concatenate text and entity features
X_hybrid = np.concatenate((X_text.toarray(), X_entities.toarray()), axis=1)

# Hierarchical clustering
agg = AgglomerativeClustering(distance_threshold=1.5, n_clusters=None)
agg.fit(X_hybrid)

cluster_labels = agg.labels_
unique_labels = set(cluster_labels)
num_clusters = len(unique_labels)

article_ids = [''] * len(news_titles)
similar_articles = [''] * len(news_titles)

cluster_first_articles = {}
for i in range(len(cluster_labels)):
    articles_in_cluster = np.where(cluster_labels == i)[0]
    cluster_num = str(i+1).zfill(3)
    for j, article_idx in enumerate(articles_in_cluster):
        article_num = str(j+1).zfill(2)
        article_id = 'SADB-' + datetime.datetime.now().strftime('%y%m%d') + '-' + cluster_num + '-' + article_num
        article_ids[article_idx] = article_id
        if j == 0:
            similar_articles[article_idx] = ','.join([x for x in article_ids if x != article_id])
            cluster_first_articles[i] = article_idx  
for i, first_article_idx in cluster_first_articles.items():
    similar_articles[first_article_idx] = ','.join([x for x in article_ids if x != article_ids[first_article_idx] and cluster_labels[article_ids.index(x)] == i])

data = pd.DataFrame({
    'title': news_titles,
    'id': article_ids,
    'cluster': cluster_labels,
    'similar_articles': similar_articles
})

data = data.sort_values(by=['cluster'])
data = data.drop('cluster', axis=1)

output = sheet.worksheet_by_title('Copy of Result')
output.set_dataframe(data, start='A1')
print('Completed successfully!')
```
